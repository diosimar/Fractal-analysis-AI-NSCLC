# Preparado de Datos solo imputando age

En este caso se trabaja con el archivo de datos sin procesar, es decir, los datos descargados directamente de la base de datos, simplemente con una columna a mayores : F.analisis, que se corresponde con la dimensión fractal calculada previamente en MATLAB y sobre el se hace la imputación de valores faltantes de edad. Se hace con edad pq se observa que la distribución es normal ycentrada por lo que al ser pocos los valores faltantes, se presupone que esta imputación no  va a tener gran impacto sobre los resultados. En cambio , el resto de valores faltantes (clinical.T.Stage , Overall.Stage, F.analysis, Histology) son datos médicos directamente relacionados con el tumor, por lo que se consideran valores más críticos para hacer imputaciones. 

```python
Data=pd.read_csv('df_sin_imputaciones.csv')
Data.head()
```

```python
Data.info()
```

## Imputación del valor de age para los datos faltantes

```python
from sklearn.impute import KNNImputer
```

```python
columnas=Data.columns
```

```python
knni = KNNImputer(n_neighbors=3)

col_imputations = knni.fit_transform(Data[['age']])
values_imputations = np.concatenate([Data[[ 'PatientID']],col_imputations,Data[[ 'clinical.T.Stage', 'Clinical.N.Stage',
       'Clinical.M.Stage', 'Overall.Stage', 'Histology', 'gender',
       'Survival.time', 'deadstatus.event', 'F.analysis']].values], axis=1)

Data2 = pd.DataFrame(values_imputations, columns=columnas)
```

```python
Data2.head()
```

```python
Data2.info()
```

```python
Data2['age']=Data2['age'].astype('float64')
Data2['clinical.T.Stage']=Data2['clinical.T.Stage'].astype('float64')
Data2['Clinical.N.Stage']=Data2['Clinical.N.Stage'].astype('int64')
Data2['Survival.time']=Data2['Survival.time'].astype('int64')
Data2['deadstatus.event']=Data2['deadstatus.event'].astype('int64')
Data2['F.analysis']=Data2['F.analysis'].astype('float64')
Data2['Clinical.M.Stage']=Data2['Clinical.M.Stage'].astype('int64')
```

```python
Data2.info()
```

## MODELO RANDOM SURVIVAL FOREST

```python
X=Datos_NoNulos[['age','clinical.T.Stage','Clinical.N.Stage','Clinical.M.Stage','F.analysis','adenocarcinoma','nos','squamous cell carcinoma','large cell','Overall.Stage','female','male']]
```

```python
X.info()
```

```python
Y_df=Datos_NoNulos[['deadstatus.event','Survival.time']]
```

```python
Y_df.info()
```

```python
Y_df['deadstatus.event']=Y_df['deadstatus.event'].astype('bool')
Y_df.info()
```

```python
Y=Y_df.to_records(index=False)
#Y
```

```python
random_state = 20

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=random_state)
```

```python
rsf = RandomSurvivalForest(
    n_estimators=1000, min_samples_split=10, min_samples_leaf=15, n_jobs=-1, random_state=random_state
)
rsf.fit(X_train, y_train)
```

```python
rsf.score(X_test, y_test)
```

```python
from sklearn.inspection import permutation_importance

result = permutation_importance(rsf, X_test, y_test, n_repeats=15, random_state=random_state)
```

```python
pd.DataFrame(
    {
        k: result[k]
        for k in (
            "importances_mean",
            "importances_std",
        )
    },
    index=X_test.columns,
).sort_values(by="importances_mean", ascending=False)
```

```python
X_test_sorted = X_test.sort_values(by=[ "F.analysis","age"])
X_test_sel = pd.concat((X_test_sorted.head(3), X_test_sorted.tail(3)))

X_test_sel
```

```python
pd.Series(rsf.predict(X_test_sel))
```
